```python
import numpy as np
import matplotlib.pyplot as plt
values = np.loadtxt('data.txt', delimiter=',', dtype=int)

values = np.log(values / np.sum(values))  # 将频率转换为概率

# 计算基本统计数据
mean = np.mean(values)  # 平均值
median = np.median(values)  # 中位数
std_dev = np.std(values)  # 标准差
minimum = np.min(values)  # 最小值
maximum = np.max(values)  # 最大值
percentile_25 = np.percentile(values, 25)  # 25%分位数
percentile_75 = np.percentile(values, 75)  # 75%分位数

# 输出结果
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
print("Minimum:", minimum)
print("Maximum:", maximum)
print("25th Percentile:", percentile_25)
print("75th Percentile:", percentile_75)

# plt.figure(figsize=(8, 6))
# plt.boxplot(values, vert=False, patch_artist=True)
# plt.title('Box and Whisker Plot of Word Frequencies')
# plt.xlabel('Frequency')
# plt.show()

counts = list(values)
plt.figure(figsize=(10, 6))
plt.hist(counts, bins=30, edgecolor='black')
plt.title('Word Frequency Distribution')
plt.xlabel('Log of Frequency')
plt.ylabel('Number of Words')
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
values = np.loadtxt('data.txt', delimiter=',', dtype=int)


plt.figure(figsize=(10, 6))
sns.kdeplot(values, bw_adjust=0.5)
plt.title('Probability Density Function of Word Frequencies')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
values = np.loadtxt('data.txt', delimiter=',', dtype=int).repeat(2)

values = np.log(values / np.sum(values))  # 将频率转换为概率

# 使用柱状图表示单词频率
plt.figure(figsize=(10, 6))
plt.bar(range(len(values)), values, color='skyblue')
plt.title('Word Frequencies')
plt.xlabel('Word Index(Unsorted)')
plt.ylabel('Log of Frequency')
# plt.xticks(word_indices, words)  # 将x轴刻度标记设置为单词名称
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
values = np.loadtxt('data.txt', delimiter=',', dtype=int).repeat(2)
values = np.sort(values)
values = np.log(values / np.sum(values))  # 将频率转换为概率

# 使用柱状图表示单词频率
plt.figure(figsize=(10, 6))
plt.bar(range(len(values)), values, color='skyblue')
plt.title('Word Frequencies')
plt.xlabel('Word Index(Sorted by Frequency)')
plt.ylabel('Log of Frequency')
# plt.xticks(word_indices, words)  # 将x轴刻度标记设置为单词名称
plt.grid(axis='y')
plt.show()
```
